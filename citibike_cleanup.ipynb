Create csv file with all months combined

```python
import pandas as pd
```

```python
october = pd.read_csv("oct21.csv")
november = pd.read_csv("nov21.csv")
december = pd.read_csv("dec21.csv")
january = pd.read_csv("jan22.csv")
february = pd.read_csv("feb22.csv")
march = pd.read_csv("mar22.csv")
april = pd.read_csv("apr22.csv")
may = pd.read_csv("may22.csv")
june = pd.read_csv("jun22.csv")
july = pd.read_csv("jul22.csv")
august = pd.read_csv("aug22.csv")
september = pd.read_csv("sep22.csv")
```

```python
oct_filtered = october.sample(10000)
nov_filtered = november.sample(10000)
dec_filtered = december.sample(10000)
jan_filtered = january.sample(10000)
feb_filtered = february.sample(10000)
mar_filtered = march.sample(10000)
apr_filtered = april.sample(10000)
may_filtered = may.sample(10000)
jun_filtered = june.sample(10000)
jul_filtered = july.sample(10000)
aug_filtered = august.sample(10000)
sep_filtered = september.sample(10000)
```

```python
year_df = pd.concat([oct_filtered, nov_filtered, dec_filtered, jan_filtered, feb_filtered, mar_filtered, apr_filtered,
                    may_filtered, jun_filtered, jul_filtered, aug_filtered, sep_filtered])
```

```python
year_df
```

```python
year_df_cleaned = year_df.dropna()
```
